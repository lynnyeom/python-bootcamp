# Debugging

the best way to debug is `print`

````python
x = True
while x :
    z = 0
    print("z = " + str(z))
    if z > 0:
        x = False
    else:
        z = 1
    print("z = " + str(z))
````

````python
x = True
z = 0
while x :
    print("z = " + str(z))
    if z > 0:
        x = False
    else:
        z = 1
````

````python
for i in range(0,50) :
    if i % 2 != 0 :
        continue
    print(i)
# ! = bang/exclamation mark
````

````python
for i in range(50) : # it should be (0 starting position ,50 end position, 2 step) 
    if i % 2 is not 0 :
        continue
    print(i)
# ! = bang/exclamation mark
````

````python
for i in range(0,50) :
    if not i % 2 == 0 :
        continue
    print(i)
# ! = bang/exclamation mark
````
